```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('insurance.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

***Features aur Target alag karo***

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

***Encoding variables***

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,LabelEncoder

le1=LabelEncoder()
le2=LabelEncoder()
X[:,1]=le1.fit_transform(X[:,1])
X[:,4]=le2.fit_transform(X[:,4])

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(drop='first'),[5])],remainder='passthrough')
X=np.array(ct.fit_transform(X))
```

```python
print(X)
```

***Train-Test Split***

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
y_train.shape
```

```python
X_train.shape
```

```python
y_test.shape
```

***Feature Scaling***

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train[:, [4, 6]] = ss.fit_transform(X_train[:, [4, 6]])
X_test[:, [4, 6]] = ss.transform(X_test[:, [4, 6]])

print(X_train)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score,mean_squared_error

print(r2_score(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```

***Polynomial Regression***

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_train_poly = poly_reg.fit_transform(X_train)
X_test_poly = poly_reg.fit_transform(X_test)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_train_poly, y_train)
```

```python
y_pred_poly=lin_reg_2.predict(X_test_poly)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred_poly.reshape(len(y_pred_poly),1), y_test.reshape(len(y_test),1)),1))
```

```python
print(r2_score(y_test, y_pred_poly))
print(np.sqrt(mean_squared_error(y_test, y_pred_poly)))
```

**SVR**

```python
from sklearn.svm import SVR
regressor_SVR = SVR(kernel = 'rbf')
regressor_SVR.fit(X_train, y_train)
```

```python
y_pred_SVR=regressor_SVR.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred_SVR.reshape(len(y_pred_SVR),1), y_test.reshape(len(y_test),1)),1))
```

```python
print(r2_score(y_test, y_pred_SVR))
print(np.sqrt(mean_squared_error(y_test, y_pred_SVR)))
```

**Desicion Tree**

```python
from sklearn.tree import DecisionTreeRegressor
regressor_DT = DecisionTreeRegressor(random_state = 0)
regressor_DT.fit(X_train, y_train)
```

```python
y_pred_DT=regressor_DT.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred_DT.reshape(len(y_pred_DT),1), y_test.reshape(len(y_test),1)),1))
```

```python
print(r2_score(y_test, y_pred_DT))
print(np.sqrt(mean_squared_error(y_test, y_pred_DT)))
```

***Random Forest***

```python
from sklearn.ensemble import RandomForestRegressor
regressor_RF = RandomForestRegressor(n_estimators = 150, random_state = 0)
regressor_RF.fit(X_train, y_train)
```

```python
y_pred_RF=regressor_RF.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred_RF.reshape(len(y_pred_RF),1), y_test.reshape(len(y_test),1)),1))
```

```python
print(r2_score(y_test, y_pred_RF))
rmse=mean_squared_error(y_test, y_pred_RF)
print(np.sqrt(rmse))
```
